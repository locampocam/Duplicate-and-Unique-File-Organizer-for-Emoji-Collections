```python
import os
import shutil

# Configurar rutas correctamente
carpeta_origen = r"C:\Users\Laura\Desktop\emogi"  # Carpeta con los emojis
carpeta_duplicados = os.path.join(carpeta_origen, "duplicados")
carpeta_no_duplicados = os.path.join(carpeta_origen, "emojis no duplicados")
archivo_lista = r"C:\Users\Laura\Desktop\duplicados\nombres.txt"  # Ruta corregida

# Verificar si el archivo nombres.txt existe
if not os.path.exists(archivo_lista):
    print(f"Error: No se encontró el archivo {archivo_lista}")
    exit()

# Leer nombres de archivos desde nombres.txt
with open(archivo_lista, "r", encoding="utf-8") as f:
    nombres_a_mover = {line.strip() for line in f}  # Eliminamos espacios en blanco

print("\nLista de nombres en nombres.txt:")
print(nombres_a_mover)  # Verificar lo que está leyendo

# Crear carpetas si no existen
os.makedirs(carpeta_duplicados, exist_ok=True)
os.makedirs(carpeta_no_duplicados, exist_ok=True)

# Mover archivos duplicados
for archivo in nombres_a_mover:
    ruta_origen = os.path.join(carpeta_origen, archivo)
    if os.path.exists(ruta_origen):  # Verificar que el archivo realmente existe
        destino = os.path.join(carpeta_duplicados, archivo)
        shutil.move(ruta_origen, destino)
        print(f"Movido duplicado: {archivo} → {carpeta_duplicados}")
    else:
        print(f"Archivo NO encontrado en {carpeta_origen}: {archivo}")

# Mover archivos no duplicados
for archivo in os.listdir(carpeta_origen):
    ruta_origen = os.path.join(carpeta_origen, archivo)
    if os.path.isfile(ruta_origen) and archivo not in nombres_a_mover:
        destino = os.path.join(carpeta_no_duplicados, archivo)
        shutil.move(ruta_origen, destino)
        print(f"Movido no duplicado: {archivo} → {carpeta_no_duplicados}")

print("\nProceso completado.")
```
